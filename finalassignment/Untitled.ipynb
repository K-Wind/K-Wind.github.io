```python
import json

jsonFile = open('nyc-hs.json', 'r')
values = json.load(jsonFile)
jsonFile.close()

# DBN = 8
# Name = 14
# Zip = 27
# Number of students = 28
# Language classes = 49
# Extracurricular = 51
# Sports - Boys = 52
# Sports - Girls = 53
# Sports - Coed = 54
# School District = 71
# Latitude = 72
# Longitude = 73

i = 0
json_data = {}
json_data["json"] = "High Schools"

data = []

for v in values["data"]:
    data.append({"DBN" : v[8],
                "Name" : v[14],
                "Zip" : v[27],
                "District" : v[71],
                "Students" : v[28],
                "Languages" : v[49],
                "Extracurricular" : v[51],
                "SportsBoys" : v[51],
                "SportsGirls" : v[52],
                "SportsCoed" : v[53],
                "Latitude" : v[72],
                "Longitude": v[73]})
json_data["data"] = data

json.dump(json_data, open("HS-data.json", "w"))
#f.write(json.dumps(data))
#f.close

jsonFile = open('nyc-sat.json', 'r')
values = json.load(jsonFile)
jsonFile.close()

# DBN = 8
# SAT Takers = 10
# Reading Average = 11
# Math Average = 12
# Writing Average = 13

i = 0
json_data = {}
json_data["json"] = "SAT"

data = []

for v in values["data"]:
    if(v[10] != "s" and v[11] != "s" and v[12] != "s" and v[13] != "s"):
        data.append({"DBN" : v[8],
                    "Takers" : v[10],
                    "Reading" : v[11],
                    "Math" : v[12],
                    "Writing" : v[13]})
json_data["data"] = data

json.dump(json_data, open("SAT-data.json", "w"))

jsonFile = open('nyc-performance.json', 'r')
values = json.load(jsonFile)
jsonFile.close()

# DBN = 8
# Graduation = 10
# College = 11

i = 0
json_data = {}
json_data["json"] = "Graduation"

data = []

for v in values["data"]:
    data.append({"DBN" : v[8],
                "Graduation" : v[10],
                "College" : v[11]})
json_data["data"] = data

json.dump(json_data, open("Grad-data.json", "w"))
```
